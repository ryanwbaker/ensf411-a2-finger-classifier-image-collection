## 1. Visualize training and validation images 
### 1.1 Create a DataBlock and DataLoaders

```python
path = Path('./digits/') ## Enter the path here
fingers = DataBlock(
    blocks=(ImageBlock, CategoryBlock), 
    get_items=get_image_files, 
    splitter=RandomSplitter(valid_pct=0.2), 
    get_y=parent_label, 
    batch_tfms=aug_transforms()) 


dls = fingers.dataloaders(path)
```

### Question 1: How many training and validation images are there for each class?
*There are 40 training and 10 validation images for each class.*

### 1.2 Show 25 training images

```python
dls.train.show_batch(max_n=25, nrows=5)
    
```

### 1.3 Show 25 validation images

```python
dls.valid.show_batch(max_n=25, nrows=5)
```
